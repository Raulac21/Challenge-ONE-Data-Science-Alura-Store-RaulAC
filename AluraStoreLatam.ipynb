### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresototal1= tienda["Precio"].sum()
print("TOTAL DE INGRESOS EN TIENDA 1 ES: ",ingresototal1)
```

```python
ingresototal2= tienda2["Precio"].sum()
print("TOTAL DE INGRESOS EN TIENDA 2 ES: ",ingresototal2)
```

```python
ingresototal3= tienda3["Precio"].sum()
print("TOTAL DE INGRESOS EN TIENDA 3 ES: ",ingresototal3)
```

```python
ingresototal4= tienda4["Precio"].sum()
print("TOTAL DE INGRESOS EN TIENDA 4 ES: ",ingresototal4)
```

**TABLA DE TIENDAS POR VENTAS**

```python
Ventas_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas Totales': [
        ingresototal1, ingresototal2, ingresototal3, ingresototal4
     ]
})
Ventas_totales
```

# 2. Ventas por categoría

```python
categoria1 = tienda["Categoría del Producto"].value_counts()
categoria2 = tienda2["Categoría del Producto"].value_counts()
categoria3 = tienda3["Categoría del Producto"].value_counts()
categoria4 = tienda4["Categoría del Producto"].value_counts()

mas_vendida1 = categoria1.idxmax()
mas_vendida2 = categoria2.idxmax()
mas_vendida3 = categoria3.idxmax()
mas_vendida4 = categoria4.idxmax()

Categoria_mas = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Categoria Mas Famosa': [
        mas_vendida1, mas_vendida2, mas_vendida3, mas_vendida4
     ]
})

Categoria_mas
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda["Calificación"].mean().round(2)
print('La Calificacion promedio de la tienda es:',promedio1)
```

```python
promedio2 = tienda2["Calificación"].mean().round(2)
print('La Calificacion promedio de la tienda es:',promedio2)
```

```python
promedio3 = tienda3["Calificación"].mean().round(2)
print('La Calificacion promedio de la tienda es:',promedio3)
```

```python
promedio4 = tienda4["Calificación"].mean().round(2)
print('La Calificacion promedio de la tienda es:',promedio4)
```

**TABLA DE CALIFIACIONES DE TIENDAS**

```python
Calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificacion de Tienda': [
        promedio1, promedio2, promedio3, promedio4
     ]
})

Calificaciones
```

# 4. Productos más y menos vendidos

```python
producto1 = tienda.groupby("Producto").size()
producto2 = tienda2.groupby("Producto").size()
producto3 = tienda3.groupby("Producto").size()
producto4 = tienda4.groupby("Producto").size()


mas_vendido1 = producto1.idxmax()
mas_vendido2 = producto2.idxmax()
mas_vendido3 = producto3.idxmax()
mas_vendido4 = producto4.idxmax()

menos_vendido1 = producto1.idxmin()
menos_vendido2 = producto2.idxmin()
menos_vendido3 = producto3.idxmin()
menos_vendido4 = producto4.idxmin()

ventas_por_producto = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Mas Vendidos': [
        mas_vendido1, mas_vendido2, mas_vendido3, mas_vendido4
     ],
    'Menos Vendidos': [
        menos_vendido1, menos_vendido2, menos_vendido3, menos_vendido4
     ]
})

ventas_por_producto
```

# 5. Envío promedio por tienda

```python
cpromedio = tienda["Costo de envío"].mean().round(2)
cpromedio2 = tienda2["Costo de envío"].mean().round(2)
cpromedio3 = tienda3["Costo de envío"].mean().round(2)
cpromedio4 = tienda4["Costo de envío"].mean().round(2)

coste_envio_prom = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo Envio Promedio': [
        cpromedio, cpromedio2, cpromedio3, cpromedio4
     ]
})

coste_envio_prom
```

***TABLA FINAL DE DATOS GENERALES DE TIENDAS***

```python
df_final = pd.concat(
    [
        Ventas_totales.set_index("Tienda"),
        Categoria_mas.set_index("Tienda"),
        Calificaciones.set_index("Tienda"),
        ventas_por_producto.set_index("Tienda"),
        coste_envio_prom.set_index("Tienda")
    ],
    axis=1
).reset_index()

df_final
```

# Graficos por tienda para su analisis

```python
import matplotlib.pyplot as plt
colors_by_shop = ['#05f03c', '#1375ec', '#ffee00', '#fc0505']
```

### Totales Por Tienda

```python
line_plot = Ventas_totales.plot(
    x='Tienda',
    y='Ventas Totales',
    kind='line',
    color='black',
    marker='o',
    figsize=(10,6),
    grid=True,
)

line_plot.scatter(
    Ventas_totales['Tienda'],
    Ventas_totales['Ventas Totales'],
    color=colors_by_shop,
    s=200,
    zorder=2
)

plt.title("Ventas Totales por Tienda")
plt.ylabel("Ventas Totales")
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

### Categorias Preferidas

```python
df_categorias = pd.concat(
    [
        categoria1.rename("Tienda 1"),
        categoria2.rename("Tienda 2"),
        categoria3.rename("Tienda 3"),
        categoria4.rename("Tienda 4")
    ],
    axis=1
).fillna(0)
```

```python
df_categorias.plot(kind='line', marker='o', figsize=(10,4), grid=True, color=colors_by_shop)

plt.title("Ventas por Categoría en Cada Tienda")
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.show()
```

### Productos mas y menos vendidos:

```python

ventas_por_producto = pd.concat(
    [
        producto1.rename('Tienda 1'),
        producto2.rename('Tienda 2'),
        producto3.rename('Tienda 3'),
        producto4.rename('Tienda 4')
    ],
    axis=1
).fillna(0)

ventas_por_producto.plot(kind='barh', color=colors_by_shop, figsize=(10,30), grid=True)
```

### Calificaciones De Tiendas

```python
plt.figure(figsize=(10,5))

plt.scatter(
    Calificaciones['Tienda'],
    Calificaciones['Calificacion de Tienda'],
    color=colors_by_shop,
    s=300,
    zorder=2
)
plt.title("Calificación Por Tienda")
plt.ylabel("Calificación")
plt.grid(True)
plt.show()
```
